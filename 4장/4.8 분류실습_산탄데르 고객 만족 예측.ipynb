- 데이터 다운로드 : https://www.kaggle.com/c/santander-customer-satisfaction
- 370개의 피처로 이루어진 데이터 셋 기반 고객만족 여부 예측
- 피처이름은 모두 익명화 되어 있어 어떤 속성인지는 알 수 없음
- 클래스 레이블명은 TARGET이며, 1(불만), 0(만족) 고객
- 모델 성능 평가는 ROC-AUC(ROC 곡선 영역)로 평가
- 대부분 만족이고 불만족인 데이터는 일부이므로 정확도 수치보다는 ROC-AUC가 더 적합

# 1. 데이터 전처리

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib

cust_df = pd.read_csv("./santander-customer-satisfaction/train.csv",encoding='latin-1')
# 데이터셋 형태 확인
print('dataset shape:', cust_df.shape)

# 데이터셋 상위 3개 컬럼 확인
cust_df.head(3)
```

```python
# 데이터 피처 타입 확인
cust_df.info()
```

```python
# 데이터 피처 Null값 확인
cust_df.isnull().sum().sum()
```

```python
# 타겟 피처의 분포 확인
print(cust_df['TARGET'].value_counts())

unsatisfied_cnt = cust_df[cust_df['TARGET'] == 1]['TARGET'].count()
total_cnt = cust_df['TARGET'].count()

print('unsatisfied 비율은 {0:.2f}%'.format((unsatisfied_cnt / total_cnt*100)))
```

```python
# 각 피처의 기초 통계량 확인
cust_df.describe( )
```

```python
# var3 피처의 분포 확인
print(cust_df['var3'].value_counts( )[:10])
```

```python
# var3 피처에서 -999999값은 116개로 데이터 비중이 가장 높은 2로 대체
# id 피처는 단숙 식별자이므로 제거
# var3 피처 값 대체 및 ID 피처 제거
cust_df['var3'].replace(-999999, 2, inplace=True)
cust_df.drop('ID',axis=1 , inplace=True)

# 피처 세트와 레이블 세트분리
X_features = cust_df.iloc[:, :-1]
y_labels = cust_df.iloc[:, -1]
print('피처 데이터 셋 형태:{0}'.format(X_features.shape))
```

```python
from sklearn.model_selection import train_test_split

# 전체 데이터 중 80%를 학습, 20%는 테스트용 데이터셋으로 분할
# 학습과 테스트 데이터 셋의 TARGET값 분포가 원본 데이터가 유사하지 않을 경우
# stratify=y_labels 코드 추가
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels,
                                                    test_size=0.2, random_state=0)
train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습 세트 Shape:{0}, 테스트 세트 Shape:{1}'.format(X_train.shape , X_test.shape))

# 비대칭 데이터 셋이므로 레이블의 분포가 원 데이터와 유사하게 분리 되었는지 확인
print('\n학습 세트 레이블 값 분포 비율')
print(y_train.value_counts()/train_cnt)

print('\n테스트 세트 레이블 값 분포 비율')
print(y_test.value_counts()/test_cnt)
```

# 2. XGBoost 모델 학습 및 하이퍼 파라미터 튜닝
- 평가 데이터 셋은 앞에서 분리한 테스트 데이터 셋 이용
- 테스트 데이터 셋을 XGBoost의 평가 데이터 셋으로 사용시 과적합 우려가 있지만, 진행 

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# n_estimators는 500으로, random state는 예제 수행 시마다 동일 예측 결과를 위해 설정. 
xgb_clf = XGBClassifier(n_estimators=500, random_state=156)

# 성능 평가 지표를 auc로, 조기 중단 파라미터는 100으로 설정하고 학습 수행. 
xgb_clf.fit(X_train, y_train, early_stopping_rounds=100,
            eval_metric="auc", eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

칼럼의 개수가 많으므로 과적합 가능성을 가정하고, max_depth, min—child_weight, colsample_bytree 하이퍼 파라미터만 일차 튜닝  
학습 시간이 많이 필요한 ML 모델 경우, 하이퍼 파라미터 튜닝을 수행하는 요령 중  
첫 번째는 먼저 2 ~ 3개 정도의 파라미터를 결합해 최적 파라미터를 찾아낸 뒤에  
이 최적 파라미터를 기반으로 다시 1~2개 파라미터를 결합해 파라미터 튜닝을 수행

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 400->100으로 감소
xgb_clf = XGBClassifier(n_estimators=100)

params = {'max_depth':[5,7], 'min_child_weight':[1,3], 'colsample_bytree':[0.5,0.75]}

# 하이퍼 파라미터 테스트의 수행속도를 향상 시키기 위해 cv 를 지정하지 않음.
# early_stopping_rounds = 100->30으로 감소
gridcv = GridSearchCV(xgb_clf, param_grid=params)
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric="auc",
           eval_set=[(X_train, y_train), (X_test, y_test)])

print('GridSearchCV 최적 파라미터:',gridcv.best_params_) 

xgb_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

ROC-AUC의 값이 0.8413-> 0.8434 성능 소폭 상승  
이를 기반으로 n_estimator를 1000으로 증가시키고, learning_rate를 0.02로 감소, reg_alhpa = 0.03 추가

```python
# n_estimators는 1000으로 증가시키고, learning_rate=0.02로 감소, reg_alpha=0.03으로 추가함. 
xgb_clf = XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02, max_depth=5,\
                        min_child_weight=1, colsample_bytree=0.75, reg_alpha=0.03)

# evaluation metric을 auc로, early stopping은 200 으로 설정하고 학습 수행. 
xgb_clf.fit(X_train, y_train, early_stopping_rounds=200, 
            eval_metric="auc",eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

ROC-AUC의 값이 0.8413-> 0.8434-> 0.8441로 이전보다 조금 더 향상

```python
# n_estimators는 1000으로 증가시키고, learning_rate=0.02로 감소, reg_alpha=0.03으로 추가함. 
xgb_clf = XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02, max_depth=7,\
                        min_child_weight=1, colsample_bytree=0.75, reg_alpha=0.03)

# evaluation metric을 auc로, early stopping은 200 으로 설정하고 학습 수행. 
xgb_clf.fit(X_train, y_train, early_stopping_rounds=200, 
            eval_metric="auc",eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

튜닝된 모델에서 각 피처 중요도 확인

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,1,figsize=(10,8))
# 상위 20개 피처만 조회
plot_importance(xgb_clf, ax=ax , max_num_features=20,height=0.4)
```

# 3. LightGBM 모델 학습과 하이퍼 파라미터 튜닝

```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier(n_estimators=500)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="auc", eval_set=evals,
                verbose=True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

XGBoost보다 학습 수행 시간이 좀 더 단축됐음을 확인 가능  
GridSearchCV로 좀 더 다양한 하이퍼 파라미터에 대한 튜닝 수행  
튜닝 대상 : num_leaves, max_depth, min_child, min_child_samples, subsample

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 200으로 감소
LGBM_clf = LGBMClassifier(n_estimators=200)

params = {'num_leaves': [32, 64],
          'max_depth':[128, 160],
          'min_child_samples':[60, 100],
          'subsample':[0.8, 1]}


# 하이퍼 파라미터 테스트의 수행속도를 향상 시키기 위해 cv를 지정하지 않음 
gridcv = GridSearchCV(lgbm_clf, param_grid=params)
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric="auc",
           eval_set=[(X_train, y_train), (X_test, y_test)])

print('GridSearchCV 최적 파라미터:', gridcv.best_params_)
lgbm_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

ROC-AUC의 값이 0.8409-> 0.8418 로 조금 성능 소폭 상승  
해당 하이퍼 파라미터를 LightGBM에 적용하고 다시 학습하여 ROC-AUC 측정 결과 도출

```python
lgbm_clf = LGBMClassifier(n_estimators=1000, num_leaves=32, sumbsample=0.8, min_child_samples=100,
                          max_depth=128)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="auc", eval_set=evals,
                verbose=True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

LightGBM의 경우 테스트 데이터 셋에서 ROC-AUC가 약 0.8417로 측정
